```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_path = 'data/Manaus.csv'
df = pd.read_csv(csv_path,encoding='iso-8859-1',sep=';')
```

```python
#lista com algumas especificações a serem filtradas
df = df[df._classificacao.eq('Confirmado')]
print("Casos confirmados antes do Filtro: ", len(df.index)) # confirmados antes do filtro


# essa filterlist é quando tiver que usar a coluna test
#filterList = ['_comorb', 'sintoma(s)*','_etnia','_profiss_saude','^(_dt_(\w){1,10})$','_origem', '_raca', '_criterio','_srag','_evolução']

# essa filterlist nao necessita da coluna test que é pra maioria dos casos
filterList = ['_comorb', 'sintoma(s)*','_etnia','_profiss_saude','^(_dt_(\w){1,10})$','_origem', '_raca', '_criterio','_srag','_evolução','tipo_teste']


for i in range(len(filterList)):
    df = df[df.columns.drop(list(df.filter(regex=filterList[i])))]
    


#observou-se que as colunas com maior numero de NaN em suas células 
# prejudicava a quantidade e qualidade dos dados
#print(df.isnull().mean() * 100)
df.dropna(inplace=True)

#1
print("Total de casos confirmados após os filtros: ", len(df.index)) # o total de dados apos o filtro de dados faltantes são de 68k

print(df['_conclusao'].value_counts())
#print(df.head())
```

```python
# print(df_positive['_sexo'].value_counts())

#2
percentRecovered = len(df[df._conclusao.eq('Recuperado')])/len(df.index)
print("Porcentagem de individuos recuperados em relação ao todo após os filtros: %.3f" % (percentRecovered*100))

#3
print("Masculino: ", len(df[df._sexo.eq('M')]))
print("Feminino ", len(df[df._sexo.eq('F')]))

#4
print("4) Média da idade dos Confirmados é: %.3f" %df['_idade'].mean())
print("4) O desvio padrão da idade dos Confirmados é: %.3f" %df['_idade'].std())
print("4) A pessoa mais jovem a contrair tem: ", df['_idade'].min(), "anos, A mais idosa possui: ", df['_idade'].max())

#5
higherRateNeighborhood = df[['_bairro','_taxa']].max()
print("5) O bairro de maior incidencia de casos é: ", higherRateNeighborhood['_bairro'])


#6
higherRecoveredNeighbor = df.groupby(['_bairro','_conclusao'])._conclusao.count().sort_values(ascending=False).iloc[0:3]
print("6) Os bairros com maior incidencia de casos recuperados são: \n", higherRecoveredNeighbor)
# print((higherRecoveredNeighbor[['_conclusao']]
# higherRecoveredNeighbor = higherRecoveredNeighbor[''.max()

# print(higherRecoveredNeighbor)

df.head()#mostra os 5 registros do topo

```
